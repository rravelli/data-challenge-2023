```python
import pandas as pd
```

```python
path = "./../../data"
x_data_test = pd.read_csv(path + "/Xtest.csv")
x_data = pd.read_csv(path + "/Xtrain_hgcGIrA.csv", sep=",")
y_data = pd.read_csv(
    path + "/Ytrain_yL5OjS4.csv", sep=",", usecols=[1]
)  # occupancy rate
```

```python
# category station
x_data["station"] = x_data["station"].astype("category")
x_data_test["station"] = x_data["station"].astype("category")

# drop date
x_data.drop("date", axis=1, inplace=True)
x_data_test.drop("date", axis=1, inplace=True)

# drop na
x_data.dropna(subset=["hour"], inplace=True)

# fill ith zeros
x_data[["p1q0", "p2q0", "p3q0", "p0q1", "p0q2", "p0q3"]] = x_data[
    ["p1q0", "p2q0", "p3q0", "p0q1", "p0q2", "p0q3"]
].fillna(0)
x_data_test[["p1q0", "p2q0", "p3q0", "p0q1", "p0q2", "p0q3"]] = x_data_test[
    ["p1q0", "p2q0", "p3q0", "p0q1", "p0q2", "p0q3"]
].fillna(0)
# hour to int
x_data["hour"] = x_data["hour"].apply(lambda x: int(x[:2]))
x_data_test["hour"] = x_data_test["hour"].apply(
    lambda x: 0 if type(x) == float else int(x[:2])
)

cat_columns = x_data.select_dtypes(["category"]).columns
cat_columns_test = x_data_test.select_dtypes(["category"]).columns
x_data[cat_columns] = x_data[cat_columns].apply(lambda x: x.cat.codes)
x_data_test[cat_columns_test] = x_data_test[cat_columns].apply(
    lambda x: x.cat.codes
)

y_data = y_data.iloc[x_data.index]

x_data_test
```

```python
from sklearn.linear_model import LinearRegression

lin = LinearRegression()
lin.fit(x_data, y_data)
lin.score(x_data, y_data)
```

```python
y = lin.predict(x_data_test)
```
